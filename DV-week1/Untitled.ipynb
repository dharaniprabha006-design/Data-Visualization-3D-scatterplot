```python
import numpy as nk
import json

```

```python
nk.random.seed(42)
total=1000
```

```python
x=nk.random.uniform(-200,200,total)
y=nk.random.uniform(-400,400,total)
z=nk.random.uniform(-100,200,total)
print(x[:3])
```

```python
points=[]
for d in range(total):
    points.append({"x":float(x[d]),
                   "y":float(y[d]),
                   "z":float(z[d])})
    
```

```python
print(points[:4])
```

```python
with open("data.json","w")as file:
    data=json.dump(points,file,indent=4)
    print(f"{total} data generated successfully")
    
```
